```python
import json
import pymongo
from pymongo import MongoClient
```

```python
# Configurações do MongoDB
MONGO_URI = 'mongodb://localhost:27017'
DATABASE_NAME = 'clash_royale'
COLLECTION_NAME = 'players'

# Função para obter dados de um jogador
def get_player_data(player_tag):
    url = f'{base_url}/players/%23{player_tag}'
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f'Erro ao buscar dados do jogador: {response.status_code}')
        return None

# Função para armazenar dados no MongoDB
def store_player_data(player_data):
    client = MongoClient(MONGO_URI)
    db = client[DATABASE_NAME]
    collection = db[COLLECTION_NAME]
    collection.insert_one(player_data)
    client.close()

# Função principal
def main():
    player_tag = 'YOUR_PLAYER_TAG'  # Substitua pelo tag do jogador que você deseja buscar
    player_data = get_player_data(player_tag)
    if player_data:
        store_player_data(player_data)
        print('Dados do jogador armazenados com sucesso!')
    else:
        print('Nenhum dado para armazenar.')

if __name__ == '__main__':
    main()
```

### Explicação do Código

1. **Configurações da API do Clash Royale**: Configure a URL base da API e o cabeçalho de autorização usando sua chave de API.
2. **Configurações do MongoDB**: Defina a URI do MongoDB e o nome do banco de dados e da coleção.
3. **Função `get_player_data`**: Faz uma solicitação GET à API do Clash Royale para obter dados de um jogador.
4. **Função `store_player_data`**: Conecta-se ao MongoDB e insere os dados do jogador na coleção especificada.
5. **Função `main`**: Obtém os dados do jogador usando o tag do jogador e armazena esses dados no MongoDB.
